```python
import pandas as pd
import ipywidgets as widgets
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import numpy as np
```

```python
a = np.random.normal(0, 0.5, 100)
b = np.random.normal(0, 0.8, 100)
c = np.random.normal(0, 1, 100)
```

```python
arr_dict = {'a': a, 'b': b, 'c': c}
def plot_scatter(x, y):
    arr_x = arr_dict[x]
    arr_y = arr_dict[y]
    trace1 = go.Scatter(
        x = arr_x,
        y = arr_y,
        mode='markers')
    py.iplot([trace1])
    
i = widgets.interactive(plot_scatter, x = ['a', 'b', 'c'], y =['a', 'b', 'c'])
```
